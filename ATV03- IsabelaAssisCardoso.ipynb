```python
# Atividade 1 - Ajuste a interpola√ß√£o  de uma curva do tipo : y = a * sin(b*x + c) usando 
# (a) interpola√ß√£o polinomial linear e c√∫bica;
# (b) interpola√ß√£o tipo "cubic spline";
# Fa√ßa a discuss√£o dos resultados
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.interpolate import UnivariateSpline
from scipy import interpolate


x = np.linspace(-2*np.pi, 2*np.pi, 50)

def f(x, a, b, c):
    return a*np.sin(b*x + c)

y = f(x, 1, 1, 0)

ruido = 0.3*np.random.normal(size = y.size)

y += ruido

inter1 = interpolate.interp1d(x, y, kind = 'linear')

inter3 = interpolate.interp1d(x, y, kind = 'cubic')

spl = UnivariateSpline(x, y)

spl.set_smoothing_factor(1.85)

xn = np.linspace(-2*np.pi, 2*np.pi, 100 )

yn1 = inter1(xn)
yn2 = inter3(xn)
yn3 = spl(xn)

plt.plot(x, y, 'o', xn, yn1, '-', xn, yn2, '--', xn, yn3, ':')
plt.legend(['data', 'linear', 'cubic', 'spline', 'nearest'], loc = 'best')
plt.show()



#Discuss√£o: Foi plotado os pontos da fun√ß√£o f(x)= sen(x) com ru√≠do criado artificialmente. A linha alaranjada
#representa a interpola√ß√£o linear para estes pontos, a linha verde tracejada a interpola√ß√£o c√∫bica e a linha
#vermelha tracejada a interpola√ß√£o do tipo "Cubic spline". Pode-se ver que, no intervalo apresentado, a interpola-
#√ß√£o que mais se aproxima da fun√ß√£o sen(x) √© a linha vermelha, n√£o tendo tantas oscila√ß√µes como as outras duas e 
#se aproximando mais da forma de onda, logo, sendo a interpola√ß√£o mais adequada para esta fun√ß√£o.
```

```python
## Atividade 2 - Ajuste a interpola√ß√£o  de uma curva do tipo :  usando y = a/( (x - b)**2 + c )
# (a) interpola√ß√£o polinomial linear e c√∫bica;
# (b) interpola√ß√£o tipo "cubic spline"; 
# Fa√ßa a discuss√£o dos resultados

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.interpolate import UnivariateSpline
from scipy import interpolate


x = np.linspace(0.1, 5, 50)

def f(x, a, b, c):
    return a/( (x - b)**2 + c )

y = f(x, 1, 0, 0)

ruido = np.random.normal(size = y.size)

y += ruido

inter1 = interpolate.interp1d(x, y, kind = 'linear')

inter3 = interpolate.interp1d(x, y, kind = 'cubic')

spl = UnivariateSpline(x, y)

spl.set_smoothing_factor(1.75)

xn = np.linspace(0.1, 5, 100 )

yn1 = inter1(xn)
yn2 = inter3(xn)
yn3 = spl(xn)

plt.plot(x, y, 'o', xn, yn1, '-', xn, yn2, '--', xn, yn3, ':')
plt.legend(['data', 'linear', 'cubic', 'spline', 'nearest'], loc = 'best')
plt.show()

#Discuss√£o: Neste caso, a fun√ß√£o plotada foi f(x) = 1/x^2, e, como pode-se ver, todos os 3 m√©todos se mostraram efi-
#cazes para interpolar a fun√ß√£o, se aproximando do comportamento desejado, com pequenas ondula√ß√µes, n√£o havendo gran-
#des diferen√ßas percept√≠veis para este caso.
```

```python
## Atividade 3 - Ajuste a interpola√ß√£o  de uma curva do tipo :  usando y = a* sin(b*x)*sqrt(x+c)
# (a) interpola√ß√£o polinomial linear e c√∫bica;
# (b) interpola√ß√£o tipo "cubic spline"; 
# Fa√ßa a discuss√£o dos resultados

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.interpolate import UnivariateSpline
from scipy import interpolate


x = np.linspace(0.1, 20, 100)

def f(x, a, b, c):
    return a*np.sin(b*x)*np.sqrt(x+c)

y = f(x, 1, 1, 0)

ruido = 0.3*np.random.normal(size = y.size)

y += ruido

inter1 = interpolate.interp1d(x, y, kind = 'linear')

inter3 = interpolate.interp1d(x, y, kind = 'cubic')

spl = UnivariateSpline(x, y)

spl.set_smoothing_factor(1.75)

xn = np.linspace(0.1, 20, 200 )

yn1 = inter1(xn)
yn2 = inter3(xn)
yn3 = spl(xn)

plt.plot(x, y, 'o', xn, yn1, '-', xn, yn2, '--', xn, yn3, ':')
plt.legend(['data', 'linear', 'cubic', 'spline', 'nearest'], loc = 'best')
plt.show()


#Discuss√£o: Neste caso, a fun√ß√£o plotada foi f(x)=sin(x)*sqrt(x) e mais uma vez, o comportamento dos 3 m√©todos foi
#semelhante, por√©m, o m√©todo "Cubic spline" se ajustou ligeiramente melhor √† curva desejada, portando, parece o mais
#adequado para a situa√ß√£o.
```
