```python
# Atividade 1 - Ajuste de uma curva do tipo : y = a * sin(b*x + c)
# Usando curve_fit from Scipy, modifique os exemplos dados na aula anterior 
# para fazer o ajuste pedido

from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

def seno(x, a, b, c):
    return a*np.sin(b*x+c)

x = np.linspace(-2*np.pi, 2*np.pi, 100)

y = seno(x, 1, 1, 0)

ruido = np.random.normal(size = y.size)

y += ruido

fig, ax = plt.subplots()

ax.scatter(x, y, color = 'r')

pars, cov = curve_fit(f = seno, xdata = x, ydata = y, p0 = [1,1,0], bounds = [-np.inf, np.inf])

res = (y-seno(x, *pars))*(y-seno(x, *pars))
aux = sum(res)
aux2 = np.sqrt(aux/res.size)

ax.plot(x, seno(x,*pars), linestyle = '--', color = 'b')
plt.show()
```

```python
# Atividade 2 - Ajuste de uma curva do tipo : y = a*/( (x - b)**2 + c )
# Usando curve_fit from Scipy, modifique os exemplos dados na aula anterior 
# para fazer o ajuste pedido

from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

def func(x, a, b, c):
    return a/( (x - b)**2 + c )

x = np.linspace(1.1, 10, 100)

y = func(x, 1, 1, 0)

ruido = 3*np.random.normal(size = y.size)

y += ruido

fig, ax = plt.subplots()

ax.scatter(x, y, color = 'r')

pars, cov = curve_fit(f = func, xdata = x, ydata = y, p0 = [0,0,0], bounds = [-np.inf, np.inf])

res = (y-func(x, *pars))*(y-func(x, *pars))
aux = sum(res)
aux2 = np.sqrt(aux/res.size)

ax.plot(x, func(x,*pars), linestyle = '--', color = 'b')
plt.show()
```

```python
# Atividade 3 - Ajuste de uma curva do tipo : y = a* sin(b*x)*sqrt(x+c)
# Usando curve_fit from Scipy, modifique os exemplos dados na aula anterior 
# para fazer o ajuste pedido

from scipy.optimize import curve_fit
import numpy as np
from numpy import sin, sqrt
import matplotlib.pyplot as plt
import matplotlib as mpl

def func(x, a, b, c):
    return a* sin(b*x)*sqrt(x+c)

x = np.linspace(1, 30, 500)

y = func(x, 1, 1, 0)

ruido = 2*np.random.normal(size = y.size)

y += ruido

fig, ax = plt.subplots()

ax.scatter(x, y, color = 'r')

pars, cov = curve_fit(f = func, xdata = x, ydata = y, p0 = [1,1,0], bounds = [-np.inf, np.inf])

res = (y-func(x, *pars))*(y-func(x, *pars))
aux = sum(res)
aux2 = np.sqrt(aux/res.size)

ax.plot(x, func(x,*pars), linestyle = '--', color = 'b')
plt.show()
```
